```python
import random

def row(count):
    emojis = ["🕯️", "🍬", "🎀", "✨", "🎁"]
    for i in range(20):
        emojis.append("🎄")
    return "".join(random.choices(emojis, k=count))

def etree(count):
    print(" " * count + "🌟")
    for i in range(count):
        print(" " * (count - i) + row(i+1))
    print(" " * count + "🟫")

etree(40)
```
